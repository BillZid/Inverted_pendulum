```python
# 绘制收敛曲线
import matplotlib.pyplot as plt
with open('results/DiscreteQIteration_th-500_dth-500_a-3/result.txt', 'r') as f:
    lines = f.readlines()
    deltas = []
    for line in lines:
        if line[0] == 'e':
            deltas.append(float(line.split(' ')[-1]))
# 绘制50迭代后的收敛曲线，横坐标为迭代次数，纵坐标为delta
# x = range(50, len(deltas))
# plt.plot(x, deltas[50:])
plt.plot(deltas)
plt.xlabel('Iteration')
plt.ylabel('Delta')
plt.show()
```

```python
# 绘制Q矩阵热力图
import numpy as np
import seaborn as sns
# q是一个三维数组，第一维是theta，第二维是dtheta，第三维是action
q = np.load('results/DiscreteQIteration_th-600_dth-600_a-7/checkpoint.npy')
# 选取theta和dtheta的范围
q = q[0:500, 0:500, :]
# 用绘制7个热力图
fig, axes = plt.subplots(1, 7, figsize=(20, 5))
for i in range(7):
    sns.heatmap(q[:, :, i], ax=axes[i], cmap='Blues')
# 横纵坐标轴上的数字不需要显示
axes[0].set_yticklabels([])
axes[0].set_xticklabels([])
axes[1].set_yticklabels([])
axes[1].set_xticklabels([])
axes[2].set_yticklabels([])
axes[2].set_xticklabels([])
axes[3].set_yticklabels([])
axes[3].set_xticklabels([])
axes[4].set_yticklabels([])
axes[4].set_xticklabels([])
axes[5].set_yticklabels([])
axes[5].set_xticklabels([])
axes[6].set_yticklabels([])
axes[6].set_xticklabels([])
# 热力图不显示坐标轴
axes[0].set_axis_off()
axes[1].set_axis_off()
axes[2].set_axis_off()
axes[3].set_axis_off()
axes[4].set_axis_off()
axes[5].set_axis_off()
axes[6].set_axis_off()
# 热力图不显示颜色盘
axes[0].collections[0].colorbar.remove()
axes[1].collections[0].colorbar.remove()
axes[2].collections[0].colorbar.remove()
axes[3].collections[0].colorbar.remove()
axes[4].collections[0].colorbar.remove()
axes[5].collections[0].colorbar.remove()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
with open('results/SARSA_th-300_dth-300_a-3/log.txt', 'r') as f:
    lines = f.readlines()
    deltas = []
    returns = []
    for line in lines:
        if line[0] == 'e':
            deltas.append(float(line.split(' ')[-4]))
            returns.append(float(line.split(' ')[-2]))
plt.plot(deltas)
plt.xlabel('Iteration')
plt.ylabel('Delta')
plt.show()
plt.plot(returns)
plt.xlabel('Iteration')
plt.ylabel('Return')
plt.show()
```
